```python
# Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
pd.set_option("display.max_columns", 100)
```

```python
def quick_profile(df, name):
    print(f"Dataset: {name}")
    print(f"Rows: {df.shape[0]:,} | Columns: {df.shape[1]}")
    display(df.head(3))
    display(df.describe(include='all').T.head(10))
```

## 1) Movies Dataset (Ratings + Revenue)

**What it is:** A movie-level dataset with critic and audience metrics.

**High-level features:**
- `Title`, `Major Genre`, `Release Date`
- `IMDB Rating`, `Rotten Tomatoes Rating`
- `Worldwide Gross`, `Production Budget`
- `Running Time min`

**Why it's interesting:** You can compare quality vs. commercial success and explore genre-level differences.

**Source:** Vega Datasets (`movies.csv`)


```python
movies_url = "https://raw.githubusercontent.com/vega/vega-datasets/master/data/movies.csv"
movies = pd.read_csv(movies_url)

# Basic cleanup for easier plotting
movies['Release Date'] = pd.to_datetime(movies['Release Date'], errors='coerce')
for col in ['Worldwide Gross', 'Production Budget', 'IMDB Rating', 'Rotten Tomatoes Rating', 'Running Time min']:
    movies[col] = pd.to_numeric(movies[col], errors='coerce')

quick_profile(movies, "Movies")
```

### Tasks for Dataset 1
1. Create a scatter plot of **Production Budget vs Worldwide Gross** and color by `Major Genre`.
2. Compare **IMDB Rating** and **Rotten Tomatoes Rating** distributions.
3. Build a chart showing which genres have the **highest median gross**.
4. Bonus: make a time trend of average movie ratings by release year.


## 2) Billboard Songs Dataset (Music Popularity)

**What it is:** Weekly Billboard Hot 100 chart data, tracking songs and artists over time.

**High-level features:**
- `song`, `performer`
- `weekid` (chart week)
- `week_position` (rank)
- `instance` (number of weeks tracked)

**Why it's interesting:** Great for temporal analysis, popularity patterns, and artist-level performance.

**Source:** TidyTuesday Billboard dataset


```python
billboard_url = "https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-09-14/billboard.csv"
billboard = pd.read_csv(billboard_url)

billboard['weekid'] = pd.to_datetime(billboard['weekid'], errors='coerce')
quick_profile(billboard, "Billboard Songs")
```

### Tasks for Dataset 2

3. For a few songs, plot **rank trajectory over time** (remember: rank 1 is best).
4. Bonus: compare how long songs stay on chart for different decades.


## 3) Yelp Reviews Dataset (Text + Ratings)

**What it is:** A sample of Yelp reviews with user votes and star ratings.

**High-level features:**
- `stars` (rating)
- `text` (review body)
- `useful`, `funny`, `cool` (community vote counts)

**Why it's interesting:** Mixes numeric and text data, enabling sentiment-style exploration and engagement analysis.

**Source:** DAT8 course public dataset (`yelp.csv`)


```python
yelp_url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/yelp.csv"
yelp = pd.read_csv(yelp_url)

# Basic text-derived feature
yelp['review_length'] = yelp['text'].astype(str).str.len()
quick_profile(yelp, "Yelp Reviews")
```

### Tasks for Dataset 3
